```python
train_set=pd.read_csv('10-11_ssd_train.csv')
valid_set=pd.read_csv('10-11_ssd_valid.csv')

train_data=train_set.iloc[:,8:20]
valid_data=valid_set.iloc[:,6:18]
train_data=np.log(abs(train_data)+1)
valid_data=np.log(abs(valid_data)+1)
```

```python
def reg_final(predictions,df,p):
    df_copy=df.copy()
    df_copy.insert(len(df_copy.columns),'predictions',predictions)
    for i in range(len(df)):
        df1=df_copy.loc[i]
        df2=df_copy[df_copy['ds']==df1['ds']]
        df2.sort_values(by="predictions" , inplace=True, ascending=True) 
        df2=df2.reset_index(drop=True)
        df3=df2[(df2['disk_id']==df1['disk_id'])&(df2['model']==df1['model'])].index.tolist()[0]
        if df3<p*len(df2):
            predictions[i]=1
        else:
            predictions[i]=0
```

```python
lgb_train = lgb.Dataset(train_data, train_set['failure_day'])  
lgb_eval = lgb.Dataset(valid_data, valid_set['failure_day'], reference=lgb_train)
evals_result = {}
params = {  
'task': 'train',  
'boosting_type': 'gbdt',                       # 设置提升类型  
'objective': 'regression',                 # 目标函数  
'metric': {'l2', 'auc'},              # 评估函数  
'num_leaves': 31,                     # 叶子节点数  
'learning_rate': 0.05,                 # 学习速率  
'feature_fraction': 0.9,             # 建树的特征选择比例  
'bagging_fraction': 0.8,              # 建树的样本采样比例  
'bagging_freq': 5,                    # k 意味着每 k 次迭代执行bagging  
'verbose': 1                          #  <0 显示致命的, =0 显示错误 (警告), >0 显示信息  
}  
  
# 7.调用LightGBM模型，使用训练集数据进行训练（拟合）  
# Add verbosity=2 to print messages while running boosting  
my_model = lgb.train(params, lgb_train, num_boost_round=5000, valid_sets=lgb_eval, early_stopping_rounds=100,evals_result=evals_result)  
  
# 8.使用模型对测试集数据进行预测  
lgb_predictions = my_model.predict(valid_data, num_iteration=my_model.best_iteration)  
```

```python
ax = lgb.plot_metric(evals_result, metric='l2')
plt.show()
```

```python
ax = lgb.plot_metric(evals_result, metric='auc')
plt.show()
```

```python
# 9.对模型的预测结果进行评判（平均绝对误差）  
reg_final(lgb_predictions,valid_set,0.5)

# print classification report
print(classification_report(valid_set['fail'], lgb_predictions))
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix=confusion_matrix(valid_set['fail'], lgb_predictions, labels=[0, 1])
sns.heatmap(conf_matrix, annot=True,fmt='f')   
plt.show()
```

```python
xgb_params = {
    'seed': 0,  
    'eta': 0.1,
    'colsample_bytree': 0.5,
    'silent': 1,
    'subsample': 0.5,
    'booster':'gbtree',
    'max_depth': 5,
    'min_child_weight': 3
}
model = XGBRegressor(objective='reg:squarederror',seed=0,eta=0.1,colsample_bytree=0.5,subsample=0.5,max_depth=5,min_child_weight=3,n_estimators=5000)
eval_set = [(train_data,train_set['failure_day']), (valid_data,valid_set['failure_day'])]

model.fit(train_data, train_set['failure_day'], early_stopping_rounds=50, eval_metric='rmse', eval_set=eval_set, verbose=True)
# make predictions for test data
y_pred = model.predict(valid_data)
xgb_predictions = [round(value) for value in y_pred]
```

```python
import numpy as np
import matplotlib.pyplot as plt
##模型训练每次的评估结果都保存在一个字典里，依据此绘图
eval_dict= model.evals_result()
val0 = eval_dict['validation_0']['rmse']
val1 = eval_dict['validation_1']['rmse']
# logger.error(cv_output)
x_list = np.linspace(20, len(val0), num=200, dtype=int, endpoint=False)
y_val0 = [val0[x] for x in x_list]
y_val1 = [val1[x] for x in x_list]
# 开始画图
plt.title('Result Analysis')
plt.plot(x_list, y_val0, color='green', label='training rmse')
plt.plot(x_list, y_val1, color='red', label='valing rmse')
plt.legend()  # 显示图例
plt.xlabel('iteration times')
plt.ylabel('rmse')
plt.show()
```
